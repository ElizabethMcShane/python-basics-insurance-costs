# U.S. Medical Insurance Costs
The goal of this project is to analyze a csv file containing data on US insurance costs using python fundamentals. These fundamentals will be used to examine the dataset and determine its potential for other use cases. 
The dataset is from Kaggle: https://www.kaggle.com/datasets/mirichoi0218/insurance

```python
# mport the necessary library
import csv


def fill_list_from_csv(fileName, colHeader):
    '''Creates a list of values from a .csv file from the column given.'''
    empty_list = []
    with open(fileName) as csv_file:
        rdr = csv.DictReader(csv_file)
        for row in rdr:
            empty_list.append(row[colHeader])
    return empty_list

age = [int(x) for x in fill_list_from_csv('insurance.csv', 'age')]
sex = fill_list_from_csv('insurance.csv', 'sex')
bmi = [float(x) for x in fill_list_from_csv('insurance.csv', 'bmi')]
children = [int(x) for x in fill_list_from_csv('insurance.csv', 'children')]
smoker = fill_list_from_csv('insurance.csv', 'smoker')
region = fill_list_from_csv('insurance.csv', 'region')
charges = [float(x) for x in fill_list_from_csv('insurance.csv', 'charges')]
```

## Summary Statistics and Counts
The following functions will be used to calculate summary statistics for the numeric data sets and generate counts for the non-numerical data.

```python
def calculate_mean(num_list):
    '''Calculates the average of a list of numbers'''
    total_sum = 0
    for num in num_list: total_sum += num
    return total_sum / len(num_list)

def calculate_variance(num_list):
    '''Calculates the variance using Bessel's correction (n - 1). 
    Taking the square root of the result will yield the sample standard deviation.'''
    mean = calculate_mean(num_list)
    deviations = [(x - mean)**2 for x in num_list]
    sum_deviations = 0
    for x in deviations: sum_deviations += x
    varience = sum_deviations / (len(num_list) -1)
    return varience

def calculate_skewness(num_list):
    '''Calculates the sample skewness given a list of values.
    A value near 0 indicates symmetry. Positive values indicate right skew, negative values indicate 
    left skew.'''
    # calculate m3
    mean = calculate_mean(num_list)
    var_cubed = [(x - mean)**3 for x in num_list]
    sum_var_cubed = 0
    for x in var_cubed: sum_var_cubed += x
    m3 = sum_var_cubed / len(num_list)
    # calculate s3
    s3 = (calculate_variance(num_list))**(2/3)
    b1 = m3/s3
    return b1

def return_counts(list_str_vals):
    '''Returns a dictionary based on distinct string values from a list as keys and the number of times they appear in the list as values.'''
    # create a set to grab distinct values
    uq_values = set()
    for item in list_str_vals: uq_values.add(item)
    # create a dictionary with the unique values as keys
    counts_dict= {}
    for item in uq_values: counts_dict[item] = 0
    for item in list_str_vals: counts_dict[item] += 1
    return counts_dict
```

## Analysis

```python
def get_summary_stats(data_list, data_list_name):
    ''' Given a list of numeric data, returns formatted summary stats'''
    mean = calculate_mean(data_list)
    variance  = calculate_variance(data_list)
    stand_dev = variance**(1/2)
    skew = calculate_skewness(data_list)
    summary = f"""Summary of {data_list_name}:
    Mean: {mean}
    Variance: {variance}
    Standard Deviation: {stand_dev}
    Skew: {skew}
    """
    return summary
 
def format_counts(data_list, data_list_name):
    '''Formats the result of the return_counts function'''
    counts_dict = return_counts(data_list)
    summary = f"Counts for {data_list_name}:\n"
    for key, value in counts_dict.items():
        summary += f'   {key}: {value}\n'
    return summary


print(get_summary_stats(age, "Age"))
print(get_summary_stats(bmi, "BMI"))
print(get_summary_stats(children, "Children"))
print(get_summary_stats(charges, "Charges"))
# counts
print(format_counts(sex, "Gender"))
print(format_counts(smoker, "Smoker"))
print(format_counts(region, "Region"))
```
